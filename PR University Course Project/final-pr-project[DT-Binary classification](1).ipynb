```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib as plt
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import LabelEncoder
```

```python
#Read and explore the cancer-data
selected_features=['id', 'diagnosis', 'radius_mean', 'texture_mean', 
                   'perimeter_mean', 'area_mean', 'smoothness_mean', 
                   'compactness_mean', 'concavity_mean', 'concave points_mean']
data= pd.read_csv('/kaggle/input/cancer-data/Cancer_Data.csv',usecols=selected_features)
data.shape
```

```python
# Splitting the data into features (X) and target label (y)
target_feature = 'diagnosis'
# Splitting the data into features (X) and target label (y)
X = data.drop(target_feature, axis=1)
y = data[target_feature]  
```

```python
#pre-processing
data_pre = data.copy()
# identifying missing values (EDA-PRE-PROCESS)                                                                                    [1]
data.isnull().sum()
#cause some of the data could contain ? which represent missing valeu
for value in data_pre.columns:
    print(value,":", sum(data_pre[value] == '?'))
```

```python
#detecting any outliers or anomalies- by using Z-score method:                                                                     [2]
import numpy as np
def detect_outliers_zscore(data, threshold=0.5):
    z_scores = np.abs((data - np.mean(data)) / np.std(data))
    outlier_indices = np.where(z_scores > threshold)[0]
    return outlier_indices

print("Outlier indices:", outlier_indices)
```

```python
#Handle outliers:                                                                                                     [---> 3]
#Removing the outliers if they are due to errors or data corruption.
import numpy as np

def remove_outliers(data, threshold=3):
    z_scores = np.abs((data - np.mean(data)) / np.std(data))
    filtered_data = data[z_scores <= threshold]
    return filtered_data

print("Filtered data:", filtered_data)
```

```python
#normalizing the numberic feature bu using Standardization:                                                                    [4]
#Standardization: Transforms the data to have zero mean and unit variance.
numericalcols=list(data.select_dtypes(exclude='object').columns)
numericalcols.pop()
numericalcols

from sklearn.preprocessing import StandardScaler
M=StandardScaler()
data_pre[numericalcols]=M.fit_transform(data_pre[numericalcols])
print("Normalized data:")
print(data_pre[numericalcols])
```

```python
#Encode categorical variables:                                                                                                       [5]
#convert the categorical features into numerical features by using label encoder
from sklearn.preprocessing import LabelEncoder

label_encoder = LabelEncoder()

#Apply  the encoding to the target labels:
#akes the original target labels (y) as input and returns the encoded version (y_encoded).
y_encoded = label_encoder.fit_transform(y)
y_encoded
```

```python
#evaluation using cross validation
from sklearn.model_selection import KFold
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import cross_val_score
from sklearn import metrics

model = DecisionTreeClassifier()

cv = 5

# Perform cross-validation
scores = cross_val_score(model, X, y, cv=cv)

# Print the cross-validation scores
print("Cross-validation scores:", scores)
print("Mean score:", scores.mean())
```

```python
#Training phase
#decision tree classifier

# Split the data into training and testing sets
X_train, X_test, y_train, y_test =train_test_split(X, y_encoded, test_size=0.3, 
                                                   random_state=42)

# Create a decision tree classifier
clf = DecisionTreeClassifier()

# Train the decision tree classifier
clf.fit(X_train, y_train)

proba = clf.predict_proba(X_test)[:, 1]

#Evaluation 

y_pred = clf.predict(X_test)

# Decode the predicted labels
y_pred_decoded = label_encoder.inverse_transform(y_pred)

# Evaluate the accuracy of the model
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
##Evaluation 
from sklearn.metrics import confusion_matrix
from sklearn.metrics import precision_recall_fscore_support 

cm = confusion_matrix(y_test, y_pred)
print(f"Confusion Matrix for {clf.__class__.__name__}:")
print(cm)
prf = precision_recall_fscore_support(y_test, y_pred, average='binary')
print(f"Precision: {prf[0]:.4f}")
print(f"Recall: {prf[1]:.4f}")
print(f"F-score: {prf[2]:.4f}")
```

```python
from sklearn import datasets
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import GridSearchCV

# Load the iris dataset

data= pd.read_csv('/kaggle/input/cancer-data/Cancer_Data.csv',usecols=selected_features)

target_feature = 'diagnosis'

X = data.drop(target_feature, axis=1)  
y = data[target_feature]  

# Define the decision tree model
clf = DecisionTreeClassifier()

# Define the hyperparameters to search over
param_grid = {
    'max_depth': [None, 20, 35],
    'min_samples_split': [6, 20, 15],
    'min_samples_leaf': [6, 4, 2]
}

# Perform grid search with cross-validation
grid_search = GridSearchCV(clf, param_grid, cv=48)
grid_search.fit(X, y)

# Print the best hyperparameters and corresponding mean cross-validated score
print("Best Hyperparameters: ", grid_search.best_params_)
print("Best Score: ", grid_search.best_score_)
```
