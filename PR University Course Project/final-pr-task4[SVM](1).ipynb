```python
#Import scikit-learn dataset library
from sklearn import datasets

#Load dataset
cancer = datasets.load_breast_cancer()
```

```python
# print the names of the 13 features
print("Features: ", cancer.feature_names)

# print the label type of cancer('malignant' 'benign')
print("Labels: ", cancer.target_names)
```

```python
# print data(feature)shape
cancer.data.shape
```

```python
# print the cancer data features (top 5 records)
print(cancer.data[0:5])
```

```python
# print the cancer labels (0:malignant, 1:benign)
print(cancer.target)
```

```python
# Import train_test_split function
from sklearn.model_selection import train_test_split

# Split dataset into training set and test set
# 70% training and 30% test
X_train, X_test, y_train, y_test = train_test_split(cancer.data, cancer.target, 
                                                    test_size=0.3,random_state=42)
```

```python
#Import svm model
from sklearn import svm

#Create a svm Classifier
clf = svm.SVC(kernel='linear') # Linear Kernel

#Train the model using the training sets
clf.fit(X_train, y_train)

#Predict the response for test dataset
y_pred = clf.predict(X_test)
```

```python
from sklearn.model_selection import cross_val_score
# c is small value to allow margin to be a big 
clf = svm.SVC(kernel='linear', C=1, random_state=42)
scores = cross_val_score(clf, X_train, y_train, cv=5)
scores
```

```python
print("%0.2f accuracy " % (scores.mean()))
```

```python
from sklearn import svm, datasets
from sklearn.model_selection import GridSearchCV

model = svm.SVC()
parameters = {'kernel': ('linear', 'rbf'), 'C': [1, 10]}

clf = GridSearchCV(model, parameters)
clf.fit(X_train, y_train)

print(clf.best_params_)
print(clf.best_score_)
```

#Evaluating the Model

```python
#Import scikit-learn metrics module for accuracy calculation
from sklearn import metrics

# Model Accuracy: how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```

```python
##Evaluation 
from sklearn.metrics import confusion_matrix
from sklearn.metrics import precision_recall_fscore_support

cm = confusion_matrix(y_test, y_pred)
print(f"Confusion Matrix for {clf.__class__.__name__}:")
print(cm)
prf = precision_recall_fscore_support(y_test, y_pred, average='binary')
print(f"Precision: {prf[0]:.4f}")
print(f"Recall: {prf[1]:.4f}")
print(f"F-score: {prf[2]:.4f}")
```
